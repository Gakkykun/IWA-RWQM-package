```python
df = pd.read_csv(
            f'./scenario-output/scenario.csv',
            header=None
        )

data_array = df.to_numpy()
data_array = data_array.T
print(data_array.shape[0], data_array.shape[1])
```

```python
target_month_id = [3,15,27,39,51] # April
# target_month_id = [6,18,30,42,54] # July
```

```python
# DO - July
plt.rcParams.update({'font.size': 12})
x = np.arange(data_array.shape[1])

fig, ax = plt.subplots(figsize=(12,6))

labels_selected = ["Base", 
      "Ss_TG_shift10dec",
      "Ss_TG_shift40dec", 
      "Ss_TG_shift70dec",
      "Ss_TG_shift10inc",
      "Ss_TG_shift50inc", 
      "Ss_TG_shift100inc",
      "Ss_KU_shift10dec",
      "Ss_KU_shift40dec",
      "Ss_KU_shift70dec",
      "Ss_KU_shift10inc",
      "Ss_KU_shift50inc",
      "Ss_KU_shift100inc",
      "ratio_G_direct_53",
      "ratio_G_direct_63"]

# data_array_selected = data_array[july,:4].tolist() + data_array[july,7:10].tolist() +\
#                     data_array[july,-2:].tolist()
# data_array_selected = [data_array[target_month_id[0], 0]]+ data_array[target_month_id[0],-2:].tolist()
data_array_selected = data_array[target_month_id[0],:].tolist()

# ax.bar(x, data_array[6,:])
ax.bar(np.arange(len(labels_selected)), data_array_selected) 
ax.set_ylabel('Concentration ($mg/L$)')
ax.set_xlabel('Condition')
ax.set_xticks(np.arange(len(labels_selected)), labels_selected, rotation=90)
ax.set_title('DO');
```
