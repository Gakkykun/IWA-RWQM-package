```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
from IPython.core.pylabtools import figsize
import os, sys
```

```python
df = pd.read_csv(
            f'./scenario-output/scenario.csv',
            header=None
        )

data_array = df.to_numpy()
data_array = data_array.T
print(data_array.shape[0], data_array.shape[1])
```
